```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
insurance =pd.read_csv("https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv")
insurance
```

```python

from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler,OneHotEncoder
from sklearn.model_selection import train_test_split
ct= make_column_transformer(
    (MinMaxScaler(),["age","bmi","children"]),
    (OneHotEncoder(handle_unknown="ignore"),["sex","smoker","region"])
)
x=insurance.drop("charges",axis=1)
y=insurance["charges"]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
ct.fit(x_train)
x_train_normal=ct.transform(x_train)
x_test_normal=ct.transform(x_test)
x_train.iloc[0]
```

```python

x_train_normal=pd.DataFrame(x_train_normal)
x_train_normal[9]
```

```python
tf.random.set_seed(42)
insurance_model=tf.keras.Sequential([
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])
insurance_model.compile(loss=tf.keras.losses.mae,
                        optimizer=tf.keras.optimizers.SGD(),
                        metrics=["mae"])

insurance_model.fit(x_train_normal,y_train, epochs=100)
```

```python
import numpy as np
def plot_decision_boundary(model,x,y):
  #dfine the axxis boundaries of the plot and create a meshgrid
  x_min,x_max=x[:,0].min()-0.1, x[:,0].max()+0.1
  y_min,y_max=x[:,0].min()-0.1, x[:,0].max()+0.1
  xx,yy=np.meshgrid(np.linspace(x_min,x_max,100),
                    np.linspace(y_min,y_max,100))
  x_in= np.c[xx.ravel(),yy.ravel()]#stack 2d arrays together
  y_pred=model.predict(x_in)
```
